```python
import torch
import torch.nn.functional as F
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
from models.model import SOPCNN
from config import config
from data.dataset import get_test_loader
```

```python
df = pd.read_csv('./results/training_log.csv')

idx_max = df['val_accuracy'].idxmax()
row_max = df.loc[idx_max]

print(f"Epoch {row_max['epoch']} had the highest accuracy of {row_max['val_accuracy']}")
```

```python
device = config.device

model_dict = torch.load('./results/model/best_model.pth', map_location = config.device)
model = SOPCNN(10)
model.load_state_dict(model_dict['model_state_dict'])

test_loader = get_test_loader(config.batch_size, config.data_path)

model.eval()
test_loss = 0
correct = 0
test_log = []
samples_per_class = 3  # Number of samples to collect per class
collected_correct = defaultdict(int)
collected_incorrect = defaultdict(int)
correct_samples = []
incorrect_samples = []

model.to(device)

with torch.no_grad():
    for data, target in test_loader:
        data, target = data.to(device), target.to(device)
        output = model(data)
        loss = F.cross_entropy(output, target, reduction='sum').item()
        test_loss += loss
        pred = output.argmax(dim=1, keepdim=True)
        probabilities = F.softmax(output, dim=1)
        confidence = probabilities.max(dim=1)[0]
        
        # Collect samples
        for idx in range(len(data)):
            true_label = target[idx].item()
            predicted_label = pred[idx].item()
            conf_score = confidence[idx].item()
            
            sample = {
                'input': data[idx].cpu(),
                'true_label': true_label,
                'predicted_label': predicted_label,
                'confidence': conf_score,
                'probabilities': probabilities[idx].cpu().numpy()
            }
            
            # Store correct predictions
            if predicted_label == true_label and collected_correct[true_label] < samples_per_class:
                correct_samples.append(sample)
                collected_correct[true_label] += 1
                
            # Store incorrect predictions
            elif predicted_label != true_label and collected_incorrect[true_label] < samples_per_class:
                incorrect_samples.append(sample)
                collected_incorrect[true_label] += 1
        correct += pred.eq(target.view_as(pred)).sum().item()
        test_log.append((loss, correct))

test_loss /= len(test_loader.dataset)
test_accuracy = 100. * correct / len(test_loader.dataset)

print(f'\nTest set: Average loss: {test_loss:.4f}, Accuracy: {test_accuracy:.4f}%\n')
```

```python
def plot_samples(samples, title):
    if not samples:  # Skip if no samples
        return
        
    plt.figure(figsize=(15, 10))
    plt.suptitle(title, fontsize=16)
    
    n_samples = len(samples)
    n_rows = (n_samples - 1) // 5 + 1
    n_cols = min(5, n_samples)
    
    for idx, sample in enumerate(samples, 1):
        plt.subplot(n_rows, n_cols, idx)
        img = sample['input'] * 0.3081 + 0.1307
        img = img.squeeze().cpu().numpy()
        plt.imshow(img, cmap='gray')
        plt.axis('off')
        title = f"True: {sample['true_label']}\nPred: {sample['predicted_label']}\nConf: {sample['confidence']:.2f}"
        plt.title(title)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nDetailed {title}:")
    print("-" * 50)
    for i, sample in enumerate(samples, 1):
        print(f"\nSample {i}:")
        print(f"True: {sample['true_label']}, Predicted: {sample['predicted_label']} (Confidence: {sample['confidence']:.4f})")
        print("Probability distribution:")
        for d, prob in enumerate(sample['probabilities']):
            print(f"  Digit {d}: {prob:.4f}")
```

```python
plot_samples(correct_samples, "Correct Predictions")
plot_samples(incorrect_samples, "Incorrect Predictions")
```
